```python
import pandas as pd
import numpy as np
```

```python
csv_urls = [
    #{"view": "gender", "url":'https://raw.githubusercontent.com/jhellingsdata/jhellingsdata.github.io/main/Data/FRED_employRate.csv'},
    {"view" : "Overall", "url":'https://raw.githubusercontent.com/jhellingsdata/jhellingsdata.github.io/main/Data/FRED_UNEMPLOY.csv'},
    {"view": "Race", "url":'https://raw.githubusercontent.com/jhellingsdata/jhellingsdata.github.io/main/Data/FRED_UNEMPLOY_race.csv'},
    {"view": "Age", "url":'https://raw.githubusercontent.com/jhellingsdata/jhellingsdata.github.io/main/Data/FRED_UNEMPLOY_age.csv'},
    {"view": "Education", "url":'https://raw.githubusercontent.com/jhellingsdata/jhellingsdata.github.io/main/Data/FRED_UNEMPLOY_educ.csv'},
    ]
```

```python
dfs = []
for csv_url in csv_urls:
    df = pd.read_csv(csv_url["url"])
    print(csv_url)
    df["View"] = csv_url["view"]
    df.rename(columns={"date": "Date",
                       "Gender": "Series",
                       "Employment" : "Value",
                       "value" : "Value",
                       "Metric": "Series"}, inplace=True)
    df = df.query("Series != 'Persons Unemployed'")
    if max(df["Value"] > 1):
        df["Value"] = df["Value"] / 100
    dfs.append(df)
df = pd.concat(dfs)
```
