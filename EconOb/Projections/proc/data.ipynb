```python
df = pd.read_excel('../Figure4.xlsx', sheet_name='Sheet2', skiprows=1)
df = df.T
df.columns = df.iloc[0]
df = df.iloc[1:]
df.reset_index(inplace=True)
df.columns = ['Date'] + list(df.columns[1:-1]) + list(["actual"])
df = df.melt(id_vars=['Date'], var_name='Series', value_name='Value')
df.dropna(subset=['Value'], inplace=True)
df['Forecast'] = df['Series'].apply(lambda x: bool( x!='actual' and  x[-1]!='a') )
df['Real_Time'] = df['Series'].apply(lambda x: bool( x[-1]=='a') )
def convert_date(x):
    try: 
        x = float(x)
        year = str(int(np.floor(x))).zfill(4)
        if len(str(x))>4:
            month = str(int(((x%1)*12)+1)).zfill(2)
        else:
            month = "01"
        return f"{year}-{month}-01"
    except:
        return None
df['Date'] = df['Date'].apply(convert_date)
df.dropna(subset=['Date'], inplace=True)
df.to_json('fig4data.json', orient='records', date_format='DOB')
```

## Fig 6

```python
df = pd.read_excel('../Fig6DataEO.xlsx')
def convert_date(x):
    try: 
        year = x[:4]
        x = x[4:]
        if "(" in x:  
            # retain only numbers 
            x = x.strip('()')
            # zfill 
            x = str(x).zfill(2)
            quarter = x
        else:
            quarter = "01"
        return f"{year}-{quarter}-01"
    except:
        return None
df['Date'] = df['Date'].apply(convert_date)
df = df.melt(id_vars=['Date'], var_name='Series', value_name='Value')
# replace ur in series with actual
df['Series'] = df['Series'].apply(lambda x: x.replace('Ur', 'actual'))
df['Furlough'] = df['Series'].apply(lambda x: bool( 'Furlough' in x ) )
df['Metric'] = df['Series'].apply(lambda x: bool( 'Econometric' in x ) )
df.to_json('fig6data.json', orient='records', date_format='DOB')
```
